```python
import pandas as pd
import numpy as np
import pandas_datareader.data as web
import matplotlib.pyplot as plt

carteira_ativos = pd.read_excel("Ativos.xlsx")
print(carteira_ativos)

```

```python
# Criando o dataframe de cotaçoes da carteira

cotacoes_carteira_ativos = pd.DataFrame()

for Ativos in carteira_ativos['Ativos']:
    cotacoes_carteira_ativos[Ativos] = web.DataReader('{}.SA'.format(Ativos), data_source='yahoo', start='2022-01-01',end='2022-06-08')['Adj Close']
    display(cotacoes_carteira_ativos)
   
```

```python
 cotacoes_carteira_ativos.info()
carteira = cotacoes_carteira_ativos / cotacoes_carteira_ativos.iloc[0]
carteira.plot(figsize=(15,5))
plt.show()
```

```python
cotacao_ibov = web.DataReader('^BVSP', data_source='yahoo', start='2022-01-01', end='2022-06-08')
display(cotacao_ibov)
```

```python
#Criando um dataframe com as quantidades de ações.
investimento = pd.DataFrame()
for ativos in carteira_ativos['Ativos']:
    investimento[ativos] = cotacoes_carteira_ativos[ativos] * carteira_ativos.loc[carteira_ativos['Ativos']==ativos, 'Qtde'].values[0]
    display(investimento)
```

```python
#Comparação entre carteira de ativos e ibov
investimento['Total'] = investimento.sum(axis=1)

investimento = investimento / investimento.iloc[0]
cotacao_ibov = cotacao_ibov / cotacao_ibov.iloc[0]

investimento['Total'].plot(figsize=(15,5), label = 'carteira_ativos')
cotacao_ibov['Adj Close'].plot(label='IBOV')
plt.legend()
plt.show()
```

```python
#Retorno do investimento
retorno_investimento =  investimento['Total'][-1] - investimento['Total'][0] - 1
retorno_ibov = cotacao_ibov['Adj Close'][-1] - cotacao_ibov['Adj Close'][0] - 1
print('Retorno da carteira: {:.2%}'.format(retorno_investimento))
print('Retorno da IBOV: {:.2%}'.format(retorno_ibov))
```

```python
#Correlação entre a carteira de ativos e IBOV.
correlacao = investimento['Total'].corr(cotacao_ibov['Adj Close'])
print(correlacao)
```
